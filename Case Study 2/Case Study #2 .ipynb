```python
import math
import numpy as np
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.stats as qq
```

## Exercise 1.4

```python
def df(n):
    set = np.linspace(-n**0.5,n**0.5)
    result = []
    
    for i in range(0, len(set)):
        
        f1=(n**(n-0.5))/(np.math.factorial(n-1))
        f2=(1+(set[i]/(n**0.5)))**(n-1)
        f3=np.exp(-n*(1+(set[i]/(n**0.5))))
        
        df=f1*f2*f3
        
        result.append(df)
        
    return set, result

mean = 0
var = 1
std = math.sqrt(var)
x = np.linspace(mean-4*std, mean+4*std, 100)

n1 = df(1)
n2 = df(2)
n10 = df(10)

plt.plot(x, stats.norm.pdf(x,mean,std),"--", label = "Normal Distribution")
plt.plot(n1[0], n1[1], "red", label = "n = 1")
plt.legend()
plt.show()

plt.plot(x, stats.norm.pdf(x,mean,std),"--", label = "Normal Distribution")
plt.plot(n2[0], n2[1], "green", label = "n = 2")
plt.legend()
plt.show()

plt.plot(x, stats.norm.pdf(x,mean,std),"--", label = "Normal Distribution")
plt.plot(n10[0], n10[1], "orange", label = "n = 10")
plt.legend()
plt.show()
```

## Exercise 2.1

```python
uni = np.random.uniform(0,1,500)

qq.probplot(uni, dist = "expon", plot=plt)

plt.ylim([0,2])

plt.title("QQ-plot\n Comparing U(0,1) to an exponential distribution")
plt.show()
```

**As shown in the graph above which we compare the exponential distribution to the observed data of a U(0,1) distribution, and we conclude that data shows a lighter right tail than the exponential distribution. As can be observed from respecitive plot below, exponential distribution has data span to positive infinity while the uniform distribution, U(0,1), involves data only up to x = 1.**

```python
# U(0,1)
size = 500

uniform = np.arange(0,1,1/size)
graph1 = np.linspace(1,1,size)

plt.plot(uniform,graph1,'b')
plt.plot((-2,0),(0,0),'b')
plt.plot((1,3),(0,0),'b')
plt.plot((0,0),(0,1),'b')
plt.plot((1,1),(1,0),'b')

plt.ylim([0,1.1])
plt.xlabel("x")
plt.ylabel("f (x)")
plt.title('U(0,1) Distribution')

plt.show()

# Exp(1)
x = np.linspace(0,5,size)
graph2 = np.exp(-x)

plt.plot(x,graph2,'g')
plt.ylim([0,1.1])
plt.plot((0,0),(0,1),'g--')
plt.xlabel("x")
plt.ylabel("f (x)")
plt.title('Exp(1) Distribution')

plt.show()
```

```python
df = pd.read_csv('IE300_CASE2_DATA', sep = " ", header = None, thousands = ",") 
df.columns = ["year", "value", "N/A"]
df = df.drop('year',1)
df = df.drop('N/A',1)
            
df = df[0:270]
logdf = np.log(df)

print(df)
print(logdf)
```

```python
log_data = []

for item in logdf["value"]:
    log_val = item
    log_data.append(log_val)

qq.probplot(log_data, dist = 'norm', plot = plt)
plt.title('QQ-plot\n comparing to Normal Distribution')
plt.show()

qq.probplot(log_data, dist = 'expon', plot = plt)
plt.title('QQ-plot\n comparing to Exponential Distribution')
plt.show()
```

**The data fits an exponential distribution as shown in the graph above, and we can conclude that because the distribution of these (270 extracted) data fits closer to the red line.**

**Furthermore, the outcome implies a considerable fit to Pareto Distribution as mentioned in the question.**
